```python
import numpy as np
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z)) # sigmoid function

def calculate_gradient(theta, X, y): # calculates gradient
    m = y.size # number of instances
    return (X.T @ sigmoid((X @ theta)-y)/m)

def gradient_descent(X, y, alpha=0.1, num_iter=100, tol=1e-7): # Does gradient descent
    X_b = np.c_[np.ones((X.shape[0], 1)), X] # adds bias/intercept

    theta = np.zeros(X_b.shape[1])

    for i in range(num_iter): # does the gradient descent
        grad = calculate_gradient(theta, X, y)
        theta -= alpha * grad

        if np.linalg.norm(grad) < tol:
            break

    return theta


def predict_proba(X, theta):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    return sigmoid(X_b @ theta)

def predict(X, theta, threshold = 0.5):
    return (predict_proba(X, theta) >= threshold).astype(int)
```
