```python
import numpy as np
```

```python
class RidgeRegression:

    def __init__(self, alpha=1.0):
        self.coef = None
        self.intercept = 0.0
        self.alpha = alpha

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)

        Xb = np.c_[np.ones((X.shape[0], 1)), X]

        penalty_dem = Xb.shape[1]
        I = np.eye(penalty_dem)
        I[0, 0] = 0.0
        
        A =np.linalg.inv(Xb.T @ Xb + alpha * I) @ Xb.T @ y

        self.intercept = A[0]
        self.coef_ = A[1:]


    def predict(self, X):
        X = np.array(X)

        return X @ self.coef_ + self.intercept_
```
